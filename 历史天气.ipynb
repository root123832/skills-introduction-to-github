```python
import pandas as pd
```

```python
df = pd.read_csv('一个月天气.csv')
df.head()
```

```python
# date = [i.split('-')[-1] for i in df['日期'].to_list()]
date = df['日期'].to_list()
h = df['最高气温'].to_list()
l = df['最低气温'].to_list()
t = df['天气'].to_list()
f = df['风向'].to_list()


```

```python
from pyecharts.charts import Pie
from pyecharts import options as opts
x = df['风向'].value_counts().index.to_list()
y = df['风向'].value_counts().to_list()

c = (
    Pie()
    .add(
        "",
        [
            list(z)
            for z in zip(
                x,
                y,
            )
        ],
        center=["40%", "50%"],
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="涿州市风向分布图"),
        legend_opts=opts.LegendOpts(type_="scroll", pos_left="80%", orient="vertical"),
    )
    .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))
    # .render("关于python招聘岗位城市分布情况.html")
)
c.render_notebook() 
```

```python
import pyecharts.options as opts
from pyecharts.charts import Line
from pyecharts.faker import Faker

c = (
    Line()
    .add_xaxis(date)
    .add_yaxis("最高气温", h)
    .add_yaxis("最低气温", l)
    .set_global_opts(title_opts=opts.TitleOpts(title="涿州2021年-1月份天气温度情况"))
    # .render("line_base.html")
)
c.render_notebook()
```

```python
# 从pyecharts库中导入options模块并简写成opts，用于设置图表的各种选项，如标题、坐标轴标签等
from pyecharts import options as opts
# 从pyecharts.charts模块中导入Bar和Line类，分别用于创建柱状图和折线图
from pyecharts.charts import Bar, Line
# 从pyecharts.faker模块中导入Faker类，用于生成虚拟数据，这里用于生成月份数据作为X轴标签
from pyecharts.faker import Faker
date = [i.split('-')[-1] for i in df['日期'].to_list()]
h = df['最高气温'].to_list()
l = df['最低气温'].to_list()
t = df['天气'].to_list()
f = df['风向'].to_list()
# 计算平均气温
average_temperature = [(high + low) / 2 for high, low in zip(h, l)]
print(average_temperature)


# 创建一个柱状图对象bar
bar = (
    Bar()
    # 使用Faker.months生成的月份数据作为X轴标签
  .add_xaxis(date)
    # 将蒸发量数据v1添加到柱状图中，系列名称为“蒸发量”
  .add_yaxis("最高气温", h)
    # 将降水量数据v2添加到柱状图中，系列名称为“降水量”
  .add_yaxis("最低气温", l)
    # 扩展Y轴，为平均温度数据创建一个新的Y轴
  .extend_axis(
        yaxis=opts.AxisOpts(
            # 设置新Y轴标签格式为 {value} °C
            axislabel_opts=opts.LabelOpts(formatter="{value} °C"),
            # 设置新Y轴刻度间隔为5
            interval=5
        )
    )
    # 设置系列选项，这里隐藏数据标签
  .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    # 设置全局选项
  .set_global_opts(
        # 设置图表标题为“Overlap-bar+line”
        title_opts=opts.TitleOpts(title="涿州市气温分布图"),
        # 设置Y轴选项，将Y轴标签格式设置为 {value} ml
        yaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(formatter="{value} °C")),
    )
)

# 创建一个折线图对象line
line = Line()
# 使用Faker.months生成的月份数据作为X轴标签
line.add_xaxis(date)
# 将平均温度数据v3添加到折线图中，并指定使用扩展的Y轴（yaxis_index=1）
line.add_yaxis("平均温度", average_temperature)

# 将折线图line重叠到柱状图bar上
bar.overlap(line)
# 在Jupyter Notebook等环境中渲染图表
bar.render_notebook()
```
